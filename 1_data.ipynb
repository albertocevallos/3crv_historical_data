```python
import pickle
import matplotlib.pyplot as plt
```

```python
# Load pickle file

data = pickle.load(open("blocks_token_logs.pkl", "rb"))
(dai, usdc, usdt) = data
```

```python
# Plot charts
plt.subplot(1, 3, 1)
plt.plot(*zip(*sorted(dai.items())))

plt.subplot(1, 3, 2)
plt.plot(*zip(*sorted(usdc.items())))

plt.subplot(1, 3, 3)
plt.plot(*zip(*sorted(usdt.items())))

plt.show()
```
